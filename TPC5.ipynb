```python
# -*- coding: utf-8 -*-
"""
Created on Sat Nov 13 20:43:15 2021

@author: MV
"""
jogo=True
p=[]
while jogo==True:
    def menu(p):
        print("""Guião de listas de inteiros:
                      (0)-Sair
                      (1)-Criar Polinómio
                      (2)-Calcular Polinómio
                      (3)-Calcular Tabela
                      (4)-Simplificar Polinómio
                      (5)-Calcular Derivada do Polinómio""")
        jog=input("Escolha uma opção:\n")    
        if jog=="0":
            global jogo
            jogo=False
            print("Terminou o programa\n")
        elif jog=="1":
            N=int(input("Quantos termos quer? "))
            for i in range(N):
                c=float(input("adicione um coeficiente: "))
                e=float(input("adicione um expoente: "))
                t=(c,e)
                p.append(t)
            print(verPol(p))
        elif jog=="2":
            x=float(input("Para que valor de x deseja calcular o polinómio? "))
            print("O valor resultante é "+str(calcPol(p,x)))
        elif jog=="3":
            nlinhas=int(input("Até que valores de x deseja calcular o polinómio? "))
            tabela(p,nlinhas)
        elif jog=="4":
            print(simplificarPol(p))
        elif jog=="5":
            x=float(input("Para que valor de x deseja calcular o polinómio? "))
            print(derivarPol(p,x))
        
    def verPol(p):
        res=str(int(p[0][0]))+"x^"+str(int(p[0][1]))
        for (c,e) in p[1:]:
            res= res+" + "+str(int(c))+"x^"+str(int(e))
        return res
    
    def calcPol(p,x):
        res=0
        for (c,e) in p:
            res= res+c*x**e
        return res
    
    def tabela(p,nlinhas):
        for i in range(nlinhas+1):
            print(str(i)+"::"+str(calcPol(p,i)))           
            
    def simplificarPol(p):            
        P1=[p[0]]
        for f in p:
            c,e=f
            if f==p[0]:
                f1=(c,e)
            elif f1[1]==e:
                n= (int(c)+int(f1[0]),e)
                P1.pop()
                P1.append(n)
                f1=(c,e)
            else:
                P1.append(f)
                f1=(c,e)
        return P1
            
    def derivarPol(p,x):
        derivada=""
        for t in p:
            c,e=t
            if int(e)==0:
                if p.index(t)==0:
                    pass
            elif int(e)==1:
                if p.index(t)==0:
                    derivada+=str(c)
                else:
                    derivada+=(" + ")+str(c)
            elif int(e)==2:
                if p.index(t)==0:
                    derivada+=str(int(e)*int(c))+"x"
                else:
                    derivada+=(" + ")+str(int(e)*int(c))+"x"
            else:
                if p.index(t)==0:
                    derivada+=str(int(e)*int(c))+"x^"+str(int(e)-1)
                else:
                    derivada+=(" + ")+str(int(e)*int(c))+"x^"+str(int(e)-1)
        return(derivada)        
    menu(p)
```
